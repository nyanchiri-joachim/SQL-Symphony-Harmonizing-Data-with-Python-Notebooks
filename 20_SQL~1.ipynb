#### Convert data type

```python
%%sql

SELECT 
    CAST(Salary AS CHAR) AS SalaryString 
FROM 
    Employees;
```

#### Trim strings

```python
%%sql

SELECT 
    TRIM(ShipName) AS TrimmedShipName 
FROM 
    Orders;
```

#### Replace string

```python
%%sql

UPDATE 
    Customers 
SET 
    ContactTitle = COALESCE(ContactTitle, 'N/A');
```

#### Extract characters

```python
%%sql
SELECT 
    SUBSTR(City, 1, 3) AS CitySubstring 
FROM 
    Customers;
```

#### String concatenation 

```python
%%sql
SELECT 
    FirstName || ' ' || LastName AS FullName 
FROM 
    Employees;
```
